```python
import matplotlib.pyplot as plt

# Data
data_samples = [1000, 1250, 1500, 1750, 2000, 2250, 2500, 2750, 3000]
rouge_1 =    [0.0435, 0.0647, 0.068, 0.0964, 0.1093, 0.1186, 0.1076, 0.1109, 0.1038]
rouge_2 =    [0.0096, 0.008,  0.012,  0.01,   0.0116, 0.0133, 0.0107, 0.0087, 0.01]
rouge_l =    [0.043,  0.0642, 0.0687, 0.096,  0.1099, 0.1188, 0.1076, 0.1108, 0.1035]
bleu =       [0.0811, 0.0764, 0.0683, 0.0869, 0.0962, 0.0955, 0.0948, 0.0992, 0.0886]
synrouge =   [0.2603, 0.2716, 0.2617, 0.3012, 0.3144, 0.3141, 0.3094, 0.3164, 0.3045]

# Plot setup
plt.figure(figsize=(14, 7))
metrics = {
    'ROUGE-1': rouge_1,
    'ROUGE-2': rouge_2,
    'ROUGE-L': rouge_l,
    'BLEU': bleu,
    'SynROUGE': synrouge
}

# Plot each metric and add text labels
for name, values in metrics.items():
    plt.plot(data_samples, values, marker='o', label=name)
    for x, y in zip(data_samples, values):
        plt.text(x, y + 0.003, f"{y:.4f}", ha='center', fontsize=8)

# Styling
plt.title("Metric Trends with Increasing Data Samples")
plt.xlabel("Number of Training Samples")
plt.ylabel("Score")
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()
plt.tight_layout()

# Save and show
plt.savefig("metric_trends.png", dpi=300)
plt.show()
```

# OOTB Gemma vs Peak Finetuned

```python
import matplotlib.pyplot as plt
import numpy as np

# Metric names
metrics = [
    "ROUGE-1", "ROUGE-2", "ROUGE-L", "BLEU",
    "BERTScore -P", "BERTScore -R", "BERTScore -F1", "SynROUGE"
]

# Values for OOTB and Fine-tuned
gemma_ootb = [0.0243, 0.006, 0.0243, 0.0205, 0.7357, 0.7026, 0.718, 0.1422]
gemma_finetuned = [0.1186, 0.0133, 0.1188, 0.0955, 0.7661, 0.7896, 0.7768, 0.3141]

# Bar settings
x = np.arange(len(metrics))
width = 0.35

fig, ax = plt.subplots(figsize=(12, 6))

# Plot bars
bars1 = ax.bar(x - width/2, gemma_ootb, width, label='Gemma 3B OOTB', color='skyblue')
bars2 = ax.bar(x + width/2, gemma_finetuned, width, label='Gemma 3B Fine-tuned (2250 Samples)', color='orange')

# Data labels
def add_labels(bars, bold=False):
    for bar in bars:
        height = bar.get_height()
        if bold:
            ax.text(bar.get_x() + bar.get_width()/2, height + 0.005,
                    f'{height:.3f}', ha='center', va='bottom', fontweight='bold')
        else:
            ax.text(bar.get_x() + bar.get_width()/2, height + 0.005,
                    f'{height:.3f}', ha='center', va='bottom')

add_labels(bars1)
add_labels(bars2, bold=True)

# Labels and title
ax.set_ylabel('Score')
ax.set_title('Gemma 3B: OOTB vs Fine-tuned (2250 Samples)')
ax.set_xticks(x)
ax.set_xticklabels(metrics, rotation=30, ha='right')
ax.legend()

plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```
